```python
import pandas as pd
import numpy as np
from datetime import datetime
from helpers import phase2_model_names as model_names
```

```python
tdf = pd.read_csv('Processed/phase2/test2.csv')
```

```python
tdf['models_min'] = tdf[model_names].min(axis=1)
tdf['models_mean'] = tdf[model_names].mean(axis=1)
tdf['models_max'] = tdf[model_names].max(axis=1)
tdf['models_std'] = tdf[model_names].std(axis=1)
tdf['models_range'] = tdf['models_max'] - tdf['models_min']
```

```python
predictions=tdf['models_mean'].to_numpy()
predictions
```

```python
ids = np.array(range(len(tdf)))+1
time = datetime.now().strftime("%d%H%M%S")
predictions_df = pd.DataFrame(ids, columns=['Id'])
predictions_df['bikes'] = predictions
predictions_df.to_csv(f'Predictions/submission_{time}.csv',index=False)
```
